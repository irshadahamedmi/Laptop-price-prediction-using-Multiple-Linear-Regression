```python
# data cleaning of unuseful columns
df.drop(['brand','Unnamed: 0','name','img_link','display(in inch)','no_of_ratings','no_of_reviews'], axis=1, inplace=True)
df.head()
```

```python
# get categorical features for lable encoding 
categorical_features =df.columns[(df.dtypes == 'object') == True].to_list()
print(categorical_features)
```

```python
#lable encoding
for feature in categorical_features:
    uniq = df[feature].unique()
    
    new_feature = []
    for el in df[feature]:
        new_feature.append(len(uniq) - np.where(uniq==el)[0][0])
    
    df[feature] = new_feature
    
df.head()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(df.drop(['price(in Rs.)'], axis=1), df['price(in Rs.)'], test_size=0.3, random_state=0)
```

```python
Model = LinearRegression().fit(X_train, Y_train)
print('r2 score: ', metrics.r2_score(Y_test, Model.predict(X_test)))
```

```python
plt.scatter(np.linspace(0, len(Model.predict(X_test)), len(Model.predict(X_test))), Model.predict(X_test), alpha=0.5, label='predictions',color='red',s=50)
plt.scatter(np.linspace(0, len(Y_test), len(Y_test)), Y_test, alpha=0.5, label='true',color='blue',marker="+",)
plt.legend()
```
